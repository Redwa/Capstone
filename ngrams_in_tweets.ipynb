```python
import nltk
import csv
import pandas as pd
from collections import Counter
from operator import itemgetter
import re
```

```python
df = pd.read_csv("tweets.csv")
```

```python
tweet = df["tweets"]
```

```python
tokens = [nltk.word_tokenize(entry.decode("utf-8")) for entry in tweet]
```

```python
bigrams = []
bigrams_enti = []
for entry in tokens:
    bigram = nltk.bigrams(entry)
    bigrams.append(bigram)
    for item in bigram:
        bigrams_enti.append(item)
        #print item
    
```

```python
trigrams = []
trigrams_enti = []
for entry in tokens:
    trigram = nltk.trigrams(entry)
    trigrams.append(trigram)
    for item in trigram:
        trigrams_enti.append(item)
        #print item
    
```

```python
trigrams_enti[:20]
```

```python
fdist_big = nltk.FreqDist(bigrams_enti)
bigramset = []
for k,v in fdist_big.items():
    bi_comp = (k,v)
    bigramset.append(bi_comp)
```

```python
bigramset
```

```python
bigramset[:30]
```

```python
sorted(bigramset,key=itemgetter(1), reverse = True)
```

```python
fdist_trig = nltk.FreqDist(trigrams_enti)
trigramset = []
for k,v in fdist_trig.items():
    tri_comp = (k,v)
    trigramset.append(tri_comp)
```
